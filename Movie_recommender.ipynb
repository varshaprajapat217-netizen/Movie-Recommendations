```python
# Now for the cast column we will create a same function who will return 3 values (actors name ) that we needed
def convert3(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L    
        
       
        
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
# for crew column we want only job that has director so for that we will make a function that fecth names of directors
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L    
            


```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

Now time to build Logic here we want our model to recommend a movie which is similar to movie that a user select now for 
that our tags simalarity should be match : For which we will convert our tags column which is in strings into vector so that 
it will easy to find simaliraty based on closest vectors rather than words

```python
# importing count vectorizer which will do our work here
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')  # to ignore stop words in tags like(and,or,for etc.)
```

```python
vector=cv.fit_transform(df['tags']).toarray()   #used fit_transform() here because I needed to both learn the vocabulary from the text data and convert it into numerical vectors in one step.
```

CountVectorizer returns a sparse matrix to save memory, so I used .toarray() to convert it into a dense array for easier use and readability.

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):  #we create a functionl called stem which convert text into list then apply porterStem
    y=[]
    for i in text.split(): #convert string (text) into list
        y.append(ps.stem(i))
    return " ".join(y)   #again converting list into " " string 
```

## Now our whole code will look like this

```python
def recommend(movie):

    # Step 1: find index of selected movie
    movie_index = df[df['title'] == movie].index[0]

    # Step 2: get similarity scores
    distance = similarity[movie_index]

    # Step 3 & 4: add index + sort by similarity (highest first)
    movie_list = sorted(list(enumerate(distance)),
                        key=lambda x: x[1],
                        reverse=True)[1:6]

    # Step 5: print similar movie indexes
    for i in movie_list:
        print(df.iloc[i[0]].title)
        
```
